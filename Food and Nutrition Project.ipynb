**<h2>Democratic and Physical Attributes</h2>**

**1. Age Distribution**

```python
sns.histplot(data=df, x='Ages', edgecolor='black', bins=20, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

**2. Height vs Weight**

```python
sns.scatterplot(data=df, x='Height', y='Weight', alpha=0.5)
plt.title('Height vs Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (Kg)')
plt.show()
```

**<br>**

**<h2>Nutritional Targets</h2>**

**1. Daily Calorie Target by Gender**

```python
sns.boxplot(data=df, x='Gender', y='Daily Calorie Target')
plt.title('Daily Calorie Target by Gender')
plt.xlabel('Gender')
plt.ylabel('Calories')
plt.show()
```

**2. Protein Intake vs Sugar Intake**

**<h5>The scatterplot of Protein Intake vs Sugar Intaake reveals dietry patterns. Clusters in high-sugar, low-protein regions indicate unhealthy diets,while balanced intakes align with better health.</h5>**

```python
sns.scatterplot(data=df, x='Protein', y='Sugar', alpha=0.5, c='blue', hue='Disease')
plt.title('Protein vs Sugar Intake')
plt.xlabel('Protein (g)')
plt.ylabel('Sugar (g)')
plt.legend(bbox_to_anchor=(1.05, 0.9))
plt.show()
```

**<h2>Activity Level and Nutrition</h2>**

**1. Calories by Activity Level**

**Analyzing Calories by Activity Level highlights energy needs based on lifestyle. Active individuals require higher calories for performance, while sedentary ones need fewer to prevent weight gain and maintain health.**

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df, x='Activity Level', y='Calories', hue='Dietary Preference', order=['Sedentary', 'Lightly Active', 'Moderately Active', 
                                                                                        'Very Active','Extremely Active'], palette='Dark2')
plt.title('Calories vs Activity Level')
plt.legend(bbox_to_anchor=(1.03,0.6))
plt.show()
```

**2. Nutrient Intake Comparison**

**The barplot comparing nutrient intake (Protein, Carbohydrates, and Fat) across different activity levels visually shows the average intake of each nutrient per activity level. It highlights how nutritional needs change with activity, providing insights into dietary adjustments for various activity levels to support better health and performance.**

```python
nutrients_melted = df.melt(id_vars='Activity Level',value_vars=['Protein','Carbohydrates','Fat'], var_name='Nutrient', value_name='Amount')
```

```python
nutrients_melted
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=nutrients_melted, x='Activity Level', y='Amount', hue='Nutrient', order=['Sedentary', 'Lightly Active', 'Moderately Active', 
                                                                                        'Very Active','Extremely Active'])
plt.title('Average Nutrient Intake by Activity Level')
plt.legend(bbox_to_anchor=(1.03,0.59))
plt.show()
```

**<h2>Disease Analysis</h2>**

**1. Disease vs Daily Calorie Target**

**The Disease vs. Daily Calorie Target plot compares the average daily calorie targets for individuals with different diseases. It helps to understand how various health conditions may affect caloric needs, offering insights into nutritional planning and potential dietary adjustments for managing specific diseases more effectively.**

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Disease', y='Daily Calorie Target', estimator='mean', hue='Gender', palette='Set2')
plt.xticks(rotation=45, ha='right',fontsize=8)
plt.title('Average Calorie Target by Disease')
plt.legend(bbox_to_anchor=(1.19,0.6))
plt.show()
```

**2. Nutritional Factors Linked to Heart Disease**

**The barplot highlighting nutritional factors linked to heart disease specifies the average intake of sodium and sugar among affected individuals. Elevated sodium and sugar levels contribute to hypertension, obesity, and cardiovascular risks. Visualizing these values underscores the importance of dietary monitoring in managing and preventing heart disease through informed nutritional choices.**

```python
# Filter the data for Heart Disease
heart_disease_data = df[df['Disease'].str.contains('Heart Disease')]

# Reshape the data for Seaborn (long format)
heart_disease_melted = heart_disease_data[['Sodium', 'Sugar']].mean().reset_index()
heart_disease_melted.columns = ['Nutrient', 'Amount']
```

```python
heart_disease_melted
```

```python
# Create the barplot using Seaborn
plt.figure(figsize=(8,6))
sns.barplot(data=heart_disease_melted, x='Nutrient', y='Amount', hue='Nutrient', palette={'Sodium':'red','Sugar':'blue'})
plt.title('Average Sodium and Sugar Intake for Heart Disease')
plt.show()
```
